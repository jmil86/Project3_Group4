```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
titanic_df = pd.read_csv('./Resources/Titanic CSV.csv')

# Display the first few rows of the dataset
print(titanic_df.head())

# Create the gradient background
plt.figure(figsize=(12, 6))
gradient = np.linspace(0, 1, 80)  # Create a vertical gradient
gradient = np.tile(gradient, (2, 1))  # Repeat for width (aligned with survival categories)
plt.imshow(gradient.T, cmap="Purples", aspect="auto", extent=(-0.5, 1.5, 1, 80))

# Overlay the boxenplot
sns.boxenplot(x='Survived', y='Age', data=titanic_df, palette='coolwarm')

# Customize the plot
plt.title('Granular Age Distribution by Survival Status (Gradient Highlight)', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Age (Granular: Individual Years)', fontsize=12)
plt.xticks([0, 1], ['Not Survived', 'Survived'], fontsize=12)

# Add finer y-axis ticks
plt.yticks(range(1, 81, 5), fontsize=10, rotation=45)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load the CSV file
titanic_df = pd.read_csv('./Resources/Titanic CSV.csv')

# Display the first few rows of the dataset
print(titanic_df.head())

# Create the boxenplot
plt.figure(figsize=(12, 6))
ax = sns.boxenplot(x='Survived', y='Age', data=titanic_df, palette='coolwarm')

# Customize the shading for the boxes
for patch in ax.artists:
    # Create a gradient colormap for the boxes
    cmap = mcolors.LinearSegmentedColormap.from_list("gradient", ["#e0f7fa", "#00695c"])
    
    # Extract the box coordinates
    x0, y0, x1, y1 = patch.get_bbox().bounds
    patch.set_facecolor(cmap(0.5))  # Set midpoint gradient color
    patch.set_alpha(0.8)  # Adjust transparency for shading

# Customize the plot labels
plt.title('Granular Age Distribution with Shaded Boxes', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.xticks([0, 1], ['Not Survived', 'Survived'], fontsize=12)

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load the CSV file
titanic_df = pd.read_csv('./Resources/Titanic CSV.csv')

# Display the first few rows of the dataset
print(titanic_df.head())

# Create the boxenplot
plt.figure(figsize=(12, 6))
ax = sns.boxenplot(x='Survived', y='Age', data=titanic_df, palette='coolwarm')

# Customize shading for the boxes based on age groups
for patch in ax.artists:
    # Extract the box coordinates
    x0, y0, x1, y1 = patch.get_bbox().bounds
    age_range = (y0, y1)

    # Dynamically apply shading: darker for 30-40 range
    if age_range[0] <= 40 and age_range[1] >= 30:
        patch.set_facecolor('#004d40')  # Dark green shade for emphasis
    else:
        patch.set_facecolor('#80cbc4')  # Light green shade for other ranges
    patch.set_alpha(0.8)  # Adjust transparency for visual effect

# Customize the plot labels
plt.title('Age Distribution with Enhanced Shading for 30-40 Age Range', fontsize=16)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.xticks([0, 1], ['Not Survived', 'Survived'], fontsize=12)

# Display the plot
plt.show()
```
